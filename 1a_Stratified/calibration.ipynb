```python
#!/usr/bin/env python3
"""
Calibration Curve Analysis for All Test Folds
Combines predictions from all outer folds and creates calibration curves
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.calibration import calibration_curve
import os
from collections import defaultdict

# Configuration
OUTPUT_ROOT = 'output'
PROBABILITIES_DIR = os.path.join(OUTPUT_ROOT, 'probabilities')
INDICES_DIR = os.path.join(OUTPUT_ROOT, 'indices')
PREDICTIONS_DIR = os.path.join(OUTPUT_ROOT, 'predictions')
DATA_PATH = 'Cancer2025exam.csv'  # Original data file

# Models to analyze (excluding SVM)
MODEL_SET = ['FS_PCA_NN', 'RF', 'FS_PCA_KNN', 'FS_PCA_QDA', 'FS_PCA_LR']

# Detect number of folds
def detect_n_folds():
    """Detect the actual number of folds from available files"""
    max_fold = 0
    fold = 1
    while os.path.exists(os.path.join(INDICES_DIR, f"test_index_{fold}.npy")):
        max_fold = fold
        fold += 1
    return max_fold

N_FOLDS = detect_n_folds()
print(f"Detected {N_FOLDS} test folds")

def load_original_data():
    """Load original data to get true labels"""
    df = pd.read_csv(DATA_PATH)
    # First column contains labels, rest are features
    y_all = df.iloc[:, 0].values
    return y_all

def load_and_combine_predictions(model_name):
    """
    Load and combine predictions from all folds for a given model
    Returns: combined sample_ids, predictions, probabilities, and true labels
    """
    all_sample_ids = []
    all_probabilities = []
    all_predictions = []
    all_true_labels = []
    
    # Load original labels
    y_original = load_original_data()
    
    # Check if model has probability files
    has_prob_files = False
    for fold_idx in range(1, N_FOLDS + 1):
        prob_path = os.path.join(PROBABILITIES_DIR, f"pred_dist_{model_name}_{fold_idx}.npy")
        if os.path.exists(prob_path):
            has_prob_files = True
            break
    
    if not has_prob_files:
        print(f"No probability files found for {model_name}")
        return None, None, None, None
    
    for fold_idx in range(1, N_FOLDS + 1):
        # Load test indices for this fold
        test_indices_path = os.path.join(INDICES_DIR, f"test_index_{fold_idx}.npy")
        test_sample_ids = np.load(test_indices_path)
        
        # Load probability predictions
        prob_path = os.path.join(PROBABILITIES_DIR, f"pred_dist_{model_name}_{fold_idx}.npy")
        if os.path.exists(prob_path):
            prob_data = np.load(prob_path)
            # First column is sample ID, rest are probabilities
            sample_ids_from_prob = prob_data[:, 0].astype(int)
            probabilities = prob_data[:, 1:]
            
            # Verify that sample IDs match
            if not np.array_equal(sample_ids_from_prob, test_sample_ids):
                print(f"Warning: Sample ID mismatch in fold {fold_idx} for {model_name}")
                # Try to align them
                aligned_probs = []
                for sid in test_sample_ids:
                    idx = np.where(sample_ids_from_prob == sid)[0]
                    if len(idx) > 0:
                        aligned_probs.append(probabilities[idx[0]])
                    else:
                        print(f"Sample {sid} not found in probability file!")
                probabilities = np.array(aligned_probs)
            
            # Load predictions to verify consistency
            pred_path = os.path.join(PREDICTIONS_DIR, f"pred_res_{model_name}_{fold_idx}.npy")
            if os.path.exists(pred_path):
                pred_data = np.load(pred_path)
                predictions = pred_data[:, 1].astype(int)
                
                # Verify predictions match max probability
                pred_from_prob = np.argmax(probabilities, axis=1) + 1  # Assuming labels start from 1
                if not np.array_equal(predictions, pred_from_prob):
                    # Check if it's just a label offset issue
                    if np.array_equal(predictions - 1, pred_from_prob - 1):
                        print(f"Note: Label offset detected and corrected for {model_name} fold {fold_idx}")
                    else:
                        print(f"Warning: Prediction mismatch in fold {fold_idx} for {model_name}")
                        print(f"Predictions from file: {predictions[:5]}")
                        print(f"Predictions from prob: {pred_from_prob[:5]}")
            
            # Get true labels for these samples
            true_labels = y_original[test_sample_ids]
            
            # Append to combined lists
            all_sample_ids.extend(test_sample_ids)
            all_probabilities.extend(probabilities)
            if 'predictions' in locals():
                all_predictions.extend(predictions)
            all_true_labels.extend(true_labels)
    
    if not all_sample_ids:
        return None, None, None, None
    
    return (np.array(all_sample_ids), 
            np.array(all_predictions) if all_predictions else None, 
            np.array(all_probabilities), 
            np.array(all_true_labels))

def plot_calibration_curves(model_name, probabilities, true_labels):
    """
    Plot calibration curves for all classes of a given model
    """
    NBins = 40  
    unique_classes = np.unique(true_labels)
    n_classes = len(unique_classes)
    
    # Determine label offset (0-based or 1-based)
    min_label = np.min(unique_classes)
    
    plt.figure(figsize=(10, 8))
    
    # Colors for different classes
    colors = plt.cm.tab10(np.linspace(0, 1, n_classes))
    
    for class_idx, class_label in enumerate(unique_classes):
        # Create One-vs-Rest binary labels
        y_binary = (true_labels == class_label).astype(int)
        
        # Get probability for this class
        # Map class label to probability index
        prob_idx = int(class_label - min_label)  # Adjust for label offset
        
        # Ensure valid probability index
        if prob_idx >= 0 and prob_idx < probabilities.shape[1]:
            y_prob = probabilities[:, prob_idx]
        else:
            print(f"Warning: Invalid probability index {prob_idx} for class {class_label}")
            continue
        
        # Clip probabilities to ensure they're in [0, 1]
        y_prob = np.clip(y_prob, 0, 1)
        
        # Calculate calibration curve with quantile strategy
        try:
            # Check if we have enough positive samples
            n_positive = np.sum(y_binary)
            n_total = len(y_binary)
            print(f"\nClass {class_label}: {n_positive} positive samples out of {n_total} total")
            
            if n_positive < 10:
                print(f"  Warning: Very few positive samples for class {class_label}")
            
            fraction_of_positives, mean_predicted_value = calibration_curve(
                y_binary, y_prob, 
                n_bins=40, 
                strategy='quantile'
            )
            
            # Plot calibration curve
            plt.plot(mean_predicted_value, fraction_of_positives, 
                    marker='o', linewidth=2, markersize=6,
                    label=f'Class {class_label}', 
                    color=colors[class_idx])
            
            # Print detailed bin information
            print(f"  Calibration bins (quantile strategy):")
            # Calculate bin edges using quantiles
            bin_edges = np.percentile(y_prob, np.linspace(0, 100, NBins + 1))
            bin_edges[0] = 0  # Ensure first edge is 0
            bin_edges[-1] = 1  # Ensure last edge is 1
            
            for i in range(len(bin_edges) - 1):
                if i < len(bin_edges) - 2:
                    bin_mask = (y_prob >= bin_edges[i]) & (y_prob < bin_edges[i+1])
                else:  # Last bin includes upper edge
                    bin_mask = (y_prob >= bin_edges[i]) & (y_prob <= bin_edges[i+1])
                
                n_samples_in_bin = np.sum(bin_mask)
                n_positive_in_bin = np.sum(y_binary[bin_mask])
                
                if n_samples_in_bin > 0:
                    actual_positive_rate = n_positive_in_bin / n_samples_in_bin
                    print(f"    Bin {i+1}: {n_samples_in_bin:4d} samples, "
                          f"{n_positive_in_bin:3d} positive "
                          f"(rate: {actual_positive_rate:.3f}, "
                          f"prob range: [{bin_edges[i]:.3f}, {bin_edges[i+1]:.3f}])")
                else:
                    print(f"    Bin {i+1}: Empty")
                
        except ValueError as e:
            print(f"Warning: Could not compute calibration for class {class_label}: {e}")
            # Fallback: try with fewer bins
            try:
                fraction_of_positives, mean_predicted_value = calibration_curve(
                    y_binary, y_prob, 
                    n_bins=5, 
                    strategy='quantile'
                )
                plt.plot(mean_predicted_value, fraction_of_positives, 
                        marker='o', linewidth=2, markersize=6,
                        label=f'Class {class_label} (5 bins)', 
                        color=colors[class_idx], linestyle='--')
                print(f"  Used 5 bins for class {class_label} due to sparse data")
            except:
                print(f"  Skipped class {class_label} - insufficient data")
    
    # Plot perfect calibration line
    plt.plot([0, 1], [0, 1], 'k--', linewidth=2, label='Perfect calibration')
    
    # Set equal aspect ratio
    plt.axis('equal')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    
    # Labels and title
    plt.xlabel('Mean Predicted Probability', fontsize=12)
    plt.ylabel('Fraction of Positives', fontsize=12)
    plt.title(f'Calibration Curves - {model_name}\n(All {N_FOLDS} test folds combined)', fontsize=14)
    
    # Grid
    plt.grid(True, alpha=0.3)
    
    # Legend - small and on the right
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
    
    # Tight layout
    plt.tight_layout()
    
    # Show plot
    plt.show()

def main():
    """
    Main analysis function
    """
    print("="*80)
    print("CALIBRATION CURVE ANALYSIS - ALL TEST FOLDS COMBINED")
    print("="*80)
    
    # Verify directories exist
    if not os.path.exists(OUTPUT_ROOT):
        print(f"Error: Output directory '{OUTPUT_ROOT}' not found!")
        return
    
    if not os.path.exists(DATA_PATH):
        print(f"Error: Original data file '{DATA_PATH}' not found!")
        return
    
    print(f"\nAnalyzing {len(MODEL_SET)} models across {N_FOLDS} test folds")
    
    # Process each model
    for model_idx, model_name in enumerate(MODEL_SET):
        print(f"\n{'='*60}")
        print(f"Processing {model_name} ({model_idx + 1}/{len(MODEL_SET)})")
        print(f"{'='*60}")
        
        try:
            # Load and combine predictions from all folds
            result = load_and_combine_predictions(model_name)
            if result[0] is None:
                print(f"Skipping {model_name} - no probability data available")
                continue
                
            sample_ids, predictions, probabilities, true_labels = result
            
            print(f"Total test samples combined: {len(sample_ids)}")
            print(f"Unique classes in true labels: {np.unique(true_labels)}")
            print(f"Probability matrix shape: {probabilities.shape}")
            
            # Verify data integrity
            # Check if predictions match max probability
            pred_from_prob = np.argmax(probabilities, axis=1)
            min_label = np.min(true_labels)
            
            # Adjust predictions from probability to match label range
            pred_from_prob_adjusted = pred_from_prob + min_label
            
            # Compare predictions
            if hasattr(predictions, '__len__'):  # Check if predictions were loaded
                mismatch_count = np.sum(predictions != pred_from_prob_adjusted)
                if mismatch_count > 0:
                    print(f"Warning: {mismatch_count} predictions don't match max probability")
                    # Show some examples
                    mismatches = np.where(predictions != pred_from_prob_adjusted)[0][:5]
                    for idx in mismatches:
                        print(f"  Sample {sample_ids[idx]}: pred={predictions[idx]}, "
                              f"max_prob_class={pred_from_prob_adjusted[idx]}")
                else:
                    print("âœ“ All predictions match maximum probability class")
            
            # Plot calibration curves
            plot_calibration_curves(model_name, probabilities, true_labels)
            
        except Exception as e:
            print(f"Error processing {model_name}: {e}")
            import traceback
            traceback.print_exc()
    
    print("\n" + "="*80)
    print("Analysis complete!")
    print("="*80)

if __name__ == "__main__":
    main()
```
